```python
df = df.drop(columns=['CRASH_DATE_EST_I', 'LANE_CNT', 'INTERSECTION_RELATED_I', 'NOT_RIGHT_OF_WAY_I', 'PHOTOS_TAKEN_I', 'STATEMENTS_TAKEN_I', 'DOORING_I', 'WORK_ZONE_I', 'WORK_ZONE_TYPE', 'WORKERS_PRESENT_I'])
df.head()
```

```python
df.HIT_AND_RUN_I
```

```python
df['HIT_AND_RUN_I'] = df['HIT_AND_RUN_I'].fillna("Z")
```

```python
df.HIT_AND_RUN_I
```

```python
num_rows_with_nan = df.isna().any(axis=1).sum()
print(num_rows_with_nan)
```

```python
df=df.dropna()
```

```python
num_rows_with_nan = df.isna().any(axis=1).sum()
print(num_rows_with_nan)
```

```python
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'])
```

```python
df['Crash Date'] = df['CRASH_DATE'].dt.date
df['Crash Time'] = df['CRASH_DATE'].dt.time
df
```

```python
df['DATE_POLICE_NOTIFIED'] = pd.to_datetime(df['DATE_POLICE_NOTIFIED'])
```

```python
df['Police Notified Date'] = df['DATE_POLICE_NOTIFIED'].dt.date
df['Police Notified Time'] = df['DATE_POLICE_NOTIFIED'].dt.time
df
```

```python
unique_values=df['TRAFFICWAY_TYPE'].unique()
print(unique_values)
```

```python
df = df.drop(columns=['CRASH_DATE', 'DATE_POLICE_NOTIFIED'])
df.head()
```

```python
df.to_csv("clean_chicago_traffic.csv", index=False)
```
